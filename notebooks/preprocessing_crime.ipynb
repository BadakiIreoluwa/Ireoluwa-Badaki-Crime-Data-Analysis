##### Combining the raw_police_data Folder into a single DataFrame

```python
# Assumes python is running from this notebook
# If your current directory is anything else change this to the relative path of the
# raw_police_data file
raw_police_data = f"{os.path.dirname(os.path.normpath(os.getcwd()))}/data/raw/raw_poilice_data"
```

```python
# Assumes you want to save the output file in the processed folder.
# If not, change this to the relative path of the folder you wish to export the processed file to
processed_folder = f"{os.path.dirname(os.path.normpath(os.getcwd()))}/data/processed"
```

```python
def parse_police_crime_data(base_folder):
    '''Parse the contents of base_folder into a dataframe and return that dataframe

    Using the os Python library we can loop through the folder to create a dataframe
    of all the csv files in the folder. This loop assumes that the raw_police_data's
    file structure has not been changed since it has been unzipped
    '''
    dataframes = []

    for folder in os.listdir(base_folder):
        folder_path = os.path.join(base_folder, folder)
        if os.path.isdir(folder_path):
            for file in os.listdir(folder_path):
                if file.endswith(".csv"):
                    file_path = os.path.join(folder_path, file)
                    df = pd.read_csv(file_path)
                    dataframes.append(df)
    
    return dataframes
```

```python
raw_crime_df = parse_police_crime_data(raw_police_data)
```

```python
crime_df = pd.concat(raw_crime_df, ignore_index=True)
crime_df.shape
crime_df.info()
crime_df.sample(10)
```

### Observations

- Context column is an empty column, Not needed for the planned analysis as crimes can be categorised by crime type
- Crime ID has many empty records. It's hard to ensure if crimes have been duplicated across months. 
- Crime ID is supposed to be a unique one way hash key to the actual crime reference number but there are many duplicate Crime IDs, these records have identical information aside from changes to Outcome Category and Crime Type and will be treated as duplicate Crime IDs.

##### Addressing Null Values and Standardising Datatypes

```python
crime_df.drop(columns=['Context'], inplace=True)
```

```python
crime_df['Month'] = pd.to_datetime(crime_df['Month'], format='%Y-%m')
```

```python
crime_df['Has_Coordinates'] = crime_df['Longitude'].notna() & crime_df["Latitude"].notna()
crime_df['Missing_Crime_ID'] = crime_df['Crime ID'].isna()
crime_df['No_Outcome'] = crime_df['Last outcome category'].isna()
```

##### Adressing Duplicates

```python
BEFORE = len(crime_df)
print(BEFORE)
```

Since the duplicates share the same month and location, the most useful record is the most **up to date** and **complete** record, hence the following deduplication rules

1. Prefer rows with a non-null Last outcome category
2. Prefer rows with location information
3. Prefer the last recorded occurence of this Crime_ID

We also need to handle crimes without Crime_IDS differently. While it is difficult to verify the uniqueness of rows with no Crime_IDs, its important to not completely discard them as there are reported and recorded crime occurences. We will use gross row deduplication as all of the crimes recorded without Crime IDs are [Anti-Social Behaviour (ASB)](https://www.met.police.uk/advice/advice-and-information/asb/asb/antisocial-behaviour/what-is-antisocial-behaviour/) reports and are not typically reported by Police Forces with a last outcome category.


```python
# Dividing the dataframe based on crimes with and without datatypes
with_id = crime_df[~crime_df['Missing_Crime_ID']]
without_id = crime_df[crime_df['Missing_Crime_ID']]
```

```python
# All crimes recorded without IDs are anti-social behaviour reports
without_id['Crime type'].value_counts()
```

```python
## How many unique crime IDs appear more than once.
duplicate_ids = with_id['Crime ID'].value_counts()
duplicate_ids = duplicate_ids[duplicate_ids > 1]

print(len(duplicate_ids))
```

```python
with_id = with_id.sort_values(
    ['Crime ID', 'No_Outcome', 'Has_Coordinates'],
    ascending=[True, False, True]
)

with_id.drop_duplicates(subset=["Crime ID"], keep='last', inplace=True)
```

```python
without_id = without_id.sort_values(['Crime ID', 'Has_Coordinates'])
without_id.drop_duplicates(keep='last', inplace=True)
```

```python
crime_df = pd.concat([with_id, without_id])
```

##### Adding Analysis Years to be used during analysis

The ONS mid-year estimates are point-in-time snapshots (mid-June each year), and for the purpose of this analysis using mid-2023 for Year 1 (data from Feb 2023 - Jan 2024) and mid-2024 for Year 2 (data from Feb 2024 - Jan 2025)

```python
def assign_year(month):
    '''Assings an analysis year to records based on the Month the crime occurred'''
    if pd.Timestamp('2023-02-01') <= month <= pd.Timestamp('2024-01-31'):
        return 'Year 1'
    elif pd.Timestamp('2024-02-01') <= month <= pd.Timestamp('2025-01-31'):
        return 'Year 2'

crime_df['analysis_year'] = crime_df['Month'].apply(assign_year)
```
